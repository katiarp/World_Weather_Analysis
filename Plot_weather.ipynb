```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import time
import numpy as np

city_data_to_load = "weather_data/cities.csv"
```

```python
city_data_df= pd.read_csv(city_data_to_load)
city_data_df.head(10)
```

```python
#Extract relevant fields from the DataFrame for plotting
lats= city_data_df["Lat"]
max_temps = city_data_df["Max Temp"]
humidity = city_data_df["Humidity"]
cloudiness = city_data_df["Cloudiness"]
wind_speed = city_data_df["Wind Speed"]
```

```python
#Get today's date in seconds.
today= time.strftime("%x")
today
```

```python
#Build the scatter plot for latitude vs. max temperature

plt.scatter(lats,
           max_temps,
           edgecolor="black",
           linewidths=1,
           marker="o",
           alpha= 0.8,
           label = "Cities")
#Incorporate the other graph properties
plt.title(f"City Latitude vs. Max Temperature " + time.strftime("%x"))
plt.ylabel("Max Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)
plt.yticks(np.arange(0,120, step=10))
#Save the figure. 
plt.savefig("weather_data/Fig1.png")

plt.show()

```

```python

#Build the scatter plot for latitude vs. humidity

plt.scatter(lats,
           humidity,
           edgecolor="black",
           linewidths=1,
           marker="o",
           alpha= 0.8,
           label = "Cities")
#Incorporate the other graph properties
plt.title(f"City Latitude vs. Humidity" + time.strftime(" %x"))
plt.ylabel("Humidity (%)")
plt.xlabel("Latitude")
plt.grid(True)
plt.yticks(np.arange(20,110, step=20))
#Save the figure. 
plt.savefig("weather_data/Fig2.png")

plt.show()
```

```python
#Build the scatter plot for latitude vs. cloudiness

plt.scatter(lats,
           cloudiness,
           edgecolor="black",
           linewidths=1,
           marker="o",
           alpha= 0.8,
           label = "Cities")
#Incorporate the other graph properties
plt.title(f"City Latitude vs. Cloudiness " + time.strftime("%x"))
plt.ylabel("Cloudiness (%)")
plt.xlabel("Latitude")
plt.grid(True)
plt.yticks(np.arange(0,110, step=20))
#Save the figure. 
plt.savefig("weather_data/Fig3.png")

plt.show()
```

```python
#Build the scatter plot for latitude vs. wind speed

plt.scatter(lats,
           wind_speed,
           edgecolor="black",
           linewidths=1,
           marker="o",
           alpha= 0.8,
           label = "Cities")
#Incorporate the other graph properties
plt.title(f"City Latitude vs. Wind Speed " + time.strftime("%x"))
plt.ylabel("Wind speed (mph)")
plt.xlabel("Latitude")
plt.grid(True)
plt.yticks(np.arange(0,50, step=10))
#Save the figure. 
plt.savefig("weather_data/Fig4.png")

plt.show()
```

```python
from scipy.stats import linregress
#Create a function to perform linear regression on the weather data
#and plot a regression line and the equation wth the data
def plot_linear_regression(x_values, y_values, title, y_label, text_coordinates):
    #Run regression on hemisphere weather data
    (slope, intercept, r_value, p_value, std_err) = linregress(x_values, y_values)
    
    #Calculate the regression line "y values" from the slope and intercept
    regress_values = x_values * slope + intercept

    # Get the equation of the line 
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

    #Create a sctter plot and plot regression line
    plt.scatter(x_values, y_values)
    plt.plot(x_values, regress_values, "r")
    #annotate the text for the line equation
    plt.annotate(line_eq, text_coordinates, fontsize=15, color = "red")
    plt.title(title)
#   plt.yticks(np.arange(40,120, step=10))
#   plt.xticks(np.arange(0,90, step=10))
    plt.xlabel("Latitud")
    plt.ylabel(y_label)
    plt.show()
```
